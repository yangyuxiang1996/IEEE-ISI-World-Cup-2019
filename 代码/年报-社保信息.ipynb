```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
data_input = pd.read_excel('E:\\Graduate_Project\\IEEE ISI World Cup 2019\\competition1\\test\\年报-社保信息.xlsx')
display(data_input.head())
```

```python
data = data_input[data_input['年报年份'].isin(['2016', '2017'])]
data.head()
```

```python
# data = data.dropna(how = 'all', thresh = 7).reset_index(drop = True)
```

```python
data = data.replace({'选择不公示': np.nan})
data = data.replace({'企业选择不公示': np.nan})
data = data.replace({'万元': np.nan})
# data = data.dropna(how='any', axis=1, thresh=3800)
data
```

```python
col_names = data.columns.tolist()
for col in col_names[1:-1]:
    print(col)
    data[col] = data[col].str.extract(r'(\d+)', expand=False)
display(data)
```

```python
data = data.loc[:,['企业编号', '城镇职工基本养老保险人数']]
data['城镇职工基本养老保险人数'] = pd.DataFrame(data['城镇职工基本养老保险人数'], dtype=float)
data
```

```python
data = data.groupby('企业编号').mean()
data = data.reset_index()
data
```

```python
data_out = pd.read_excel('E:\\Graduate_Project\\IEEE ISI World Cup 2019\\competition1\\赛题1结果_团队名.xlsx')
data_out = pd.merge(data_out, data, how='left')
data_out.head()
```

```python
data_out.isna().sum()
```

```python
data_out.fillna(data_out.median(), inplace=True)
data_out.head()
```

```python
data_out.to_excel('年报-社保信息_特征.xlsx', index=False)
```
