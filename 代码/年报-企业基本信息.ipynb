```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
data_input = pd.read_excel('E:\\Graduate_Project\\IEEE ISI World Cup 2019\\competition1\\test\\年报-企业基本信息.xlsx')
display(data_input.head())
```

```python
data = data_input[['企业编号', '是否有网站或网点', '企业是否有投资信息或购买其他公司股权', 
                  '有限责任公司本年度是否发生股东股权转', '是否提供对外担保', '年报年份']]
data = data.dropna(how='any')
data = data[data['年报年份']==2017.0].reset_index(drop=True)
display(data)
```

```python
data = data.replace({'是':1, '有':1, '否':0, '无':0})
# data.describe()
```
