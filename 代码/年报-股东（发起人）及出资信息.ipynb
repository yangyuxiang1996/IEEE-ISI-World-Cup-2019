```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
data_input = pd.read_excel('E:\\Graduate_Project\\IEEE ISI World Cup 2019\\competition1\\test\\年报-股东（发起人）及出资信息.xlsx')
display(data_input.head())
```

```python
data_feature = data_input.loc[:,['企业编号', '实缴出资信息']]
display(data_feature.head())
```

```python
data_feature = pd.concat([data_feature, data_feature['实缴出资信息'].str.split(r'\n', expand=True)], axis=1)
display(data_feature)
# data = data_feature.loc[:, ['企业编号', '1']]
# display(data)
```

```python
data = data_feature.loc[:, ['企业编号', 1]]
data.rename(columns={1: '实缴出资额'}, inplace=True)
display(data)
```

```python
data['实缴出资额'] = data['实缴出资额'].str.extract(r'(\d+)', expand=True)
data['实缴出资额'] = pd.DataFrame(data['实缴出资额'], dtype=np.float)
data['实缴出资额']
```

```python
data = data.groupby('企业编号').mean()
```

```python
data = data.reset_index()
```

```python
data.head()
```

```python
data_out = pd.read_excel('E:\\Graduate_Project\\IEEE ISI World Cup 2019\\competition1\\赛题1结果_团队名.xlsx')
data_out = pd.merge(data_out, data, how='left')
data_out.head()
```

```python
data_out.isna().sum()
```

```python
data_out.fillna(data_out.median(), inplace=True)
data_out.isna().sum()
```

```python
data_out.head()
```

```python
data_out.to_excel('年报-股东（发起人）及出资信息.xlsx', index=False)
```
