```python
import pandas as pd
import numpy as np
import os
```

```python
# data = pd.read_csv('/home/yang/Desktop/competitionOne/yyx/01.csv')
data = pd.read_excel('E:\\Graduate_Project\\IEEE ISI World Cup 2019\\competition1\\test\\工商基本信息表.xlsx')
print(data.head())
col_names = data.columns.tolist()
print(col_names)
```

```python
index = data.index
data[data['注册资本币种(正则)'] == '美元']
```

```python
data.loc[data['注册资本币种(正则)'] == '美元', '注册资本（万元）'] = round(data.loc[data['注册资本币种(正则)'] == '美元', '注册资本（万元）']*6.72, 2)
del data['注册资本币种(正则)']
print(data['注册资本（万元）'][1309])
```

```python
data['成立日期'] = pd.to_datetime(data['成立日期']) #将object转换成日期格式
# data['成立日期'].dt.year 

duration = 2019 - data['成立日期'].dt.year    # 取年份差值
data['成立日期'] = duration
data.rename(columns={'成立日期':'成立时长'}, inplace=True) #修改列名
print(data.head())
```

```python
states = [0 for i in range(len(index))]
for i in range(len(index)):
    if ('开业' in data['经营状态'][i] or '存续' in data['经营状态'][i]):
        states[i] = 1
    elif '在业' in data['经营状态'][i]:
        states[i] = 2
    elif '迁出' in data['经营状态'][i]:
        states[i] = 3
states = pd.Series(states)
data['经营状态'] = states
print(data['经营状态'].head())
```

```python
li = [chr(i) for i in range(ord("A"),ord("R")+1)]
for i in range(len(li)):
    data.loc[data['行业大类（代码）'] == li[i], '行业大类（代码）'] = i+1
data['行业大类（代码）'].head()
```

```python
type = ['民营', '外企']
for i in range(len(type)):
    data.loc[data['类型'] == type[i], '类型'] = i+1
data['类型'].head()
```

```python
data.replace('Y', 1, inplace = True)
data.replace('N', 0, inplace = True)
data['是否上市'].head()
```

```python
data.drop(['经营期限自', '经营期限至', '发照日期'], axis=1, inplace=True)
data.head()
```

```python
data.loc[data['注销原因'].isnull(), '注销原因'] = 0
data.loc[data['注销原因'] == '决议解散', '注销原因'] = 1
data.loc[data['注销原因'] == '宣告破产', '注销原因'] = 2
data.loc[data['注销原因'] == '因公司合并或分立', '注销原因'] = 3
data.loc[data['注销原因'] == '章程规定的解散事由出现', '注销原因'] = 4
data.loc[data['注销原因'] == '其他原因', '注销原因'] = 5
data['注销原因'].head()
```

```python
data.loc[data['注销时间'].notnull(), '注销时间'] = 1
data.loc[data['注销时间'].isnull(), '注销时间'] = 0
```

```python
data = data.drop(['注册资本币种(正则)', '城市代码', '地区代码', '登记机关区域代码'], axis=1)
data.head()
```

```python
data[['行业大类（代码）', '行业小类（代码）']] = data[['行业大类（代码）', '行业小类（代码）']].fillna(data[['行业大类（代码）', '行业小类（代码）']].mode().iloc[0])
data['员工人数'] = data['员工人数'].fillna(data['员工人数'].median())
data.isna().sum()
```

```python
data['注册资本（万元）'].max()
```

```python
max_min_scalr = lambda x: (x-x.min())/(x.max()-x.min())
data[['注册资本（万元）']] = data[['注册资本（万元）']].apply(max_min_scalr)
data.head()
```
